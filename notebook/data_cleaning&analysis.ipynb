# Cleaning

```python
import pandas as pd

df = pd.read_csv(
    "superstore_final_dataset.csv",
    encoding="latin1"
)

df.head()
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
df.columns
```

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    dayfirst=True,
    errors="coerce"
)

df["ship_date"] = pd.to_datetime(
    df["ship_date"],
    dayfirst=True,
    errors="coerce"
)
```

```python
df[["order_date", "ship_date"]].isnull().sum()
```

```python
df.dtypes
```

```python
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.to_period("M").astype(str)

# Orders_per_customer metric
orders_per_customer = df.groupby("customer_id")["order_id"].nunique()
df["orders_per_customer"] = df["customer_id"].map(orders_per_customer)
```

```python
df[["order_date", "year", "month", "sales", "orders_per_customer"]].head()
```

```python
df.describe()[["sales"]]
```

```python
df.to_csv("clean_superstore.csv", index=False)
```

# Analysis

```python
df = pd.read_csv("clean_superstore.csv")
df["order_date"] = pd.to_datetime(df["order_date"])

monthly_sales = (
    df
    .groupby(df["order_date"].dt.to_period("M"))["sales"]
    .sum()
    .reset_index()
)

monthly_sales["order_date"] = monthly_sales["order_date"].astype(str)
monthly_sales.head()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Converting to time series
ts = monthly_sales.set_index("order_date")["sales"]

# Training the model
model = ExponentialSmoothing(
    ts,
    trend="add",
    seasonal=None
).fit()

# Forecasting
forecast = model.forecast(12) # Next 6 months
forecast
```

```python
forecast_df = forecast.reset_index()
forecast_df.columns = ["month", "forecasted_sales"]

forecast_df.to_csv("sales_forecast.csv", index=False)
```
